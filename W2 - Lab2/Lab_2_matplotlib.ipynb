# Matplotlib for visualisation

### What is matplotlib?

 * Matplotlib is a Python library for generating 2D and 3D plots.
 * Matlab-like plotting library and follows the naming convention of MATLAB.
 * Very popular and flexible. All aspects of a plot can be controlled programmatically.
 * Can be used to draw graphs, barcharts, scatter plots, contour and surface plots etc.
 * Can produce very high quality output in numerous formats (png, pdf, svg etc).
 
 * While we will not be covering the Seaborn library in this module, do keep in mind this is another visualisation package that makes use of matplotlib and produces excellent plots. For more details, visit seaborn.pydata.org
 
### Official website: http://matplotlib.org/

```python
# First thing to do: import the necessary libraries
# STEP 1

import matplotlib.pyplot as plt     # plt is an alias that can now be used to access the functionalities of pyplot
import numpy as np
```

# Examples

## Plotting Numpy function

```python
# The most important function in matplotlib is plot, which allows you to plot 2D data. Here is a simple example:
# Compute the x and y coordinates for points on a sine curve

#STEP2 : prepare the data for plotting
    
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

print("Printing x")
print(x)
print("Prining y")
print(y)
```

```python
# Step 3: plot the points by calling the plot function and passing the necessary parameters

plt.plot(x, y, 'b')    #x,y represents 2D data
                       #third parameter 'b' is used to chose the color


# Step 4: show the plots by calling show() function.
plt.show() # the x-axis represents the values of x, and the y-axis the values of y.             
```

## Controlling the line attributes

```python
# Playing around with: color, linewidth, linestyle parameters of the plot() function

# blue color with two pixels linewidth
y_data=np.sin(x)
plt.plot(x,y_data,color='blue',linewidth=2.0, linestyle='-')
     
# red color with two pixel linewidth and dot separated lines
y_data=np.cos(x)     
plt.plot(x,y_data,color='red',linewidth=2.0, linestyle='-.')
     
# plots grid behind the lines
plt.grid(True)

# note that we called the plot function twice here, and both lines are plotted on the same figure
```

### Adding labels, constraining the plotting range

```python
# We can use xlim(x1, x2), ylim(y1,y2) to constrain the plot range

x=np.arange(-5,5,0.1) 
y=x**2

# print the data x and y. Uncomment if you wish to see how your x and y looks like
print("Printing x")
print(x)
print("Printing y")
print(y)

plt.xlim(-5,5)   # sets the bounds of the x-axis to -5 and 5
plt.ylim(0,30)  # sets the bounds of the y-axis to 0 and 30 

plt.xlabel("x")     # Set the x axis label
plt.ylabel("y=x*x") # Set the y axis label
plt.title("Plot y=x*x") # Set the title of the plot

plt.plot(x,y)
plt.show()
```

## Set the ticks of axes

```python
# We can specify the ticks of the axes - xticks and yticks

#set the data
x=np.linspace(-np.pi,np.pi,100)
y_data = np.cos(x)

plt.xticks(np.linspace(-np.pi,np.pi,5)) # Note pi=3.14; so it produces 5 ticks between -3.14 and 3.14 (see x-axis)

plt.plot(x,y_data,color='red',linewidth=2.0,linestyle='-')
         
plt.show()
```

## Set the legends

```python
# You can add legends with the label parameter

x=np.linspace(-np.pi,np.pi,100)
#print(x)
y_sin,y_cos = np.sin(x), np.cos(x)     # Note Python allows to write two statements in a single line

plt.plot(x,y_sin,color='blue',linewidth=2.0, linestyle='-',label='sin') # The label here is set to 'sin'
plt.plot(x,y_cos,color='red',linewidth=2.0, linestyle='-',label='cos') # The label here is set to 'cos'

plt.title('Sine and Cosine')
plt.legend(loc='upper left') # Plot the legends on the upper left area of the plot
plt.show()
```

# Subplots

 * We can separate the data, on the same figure, using the subplot() function.
 * Takes three integer parameters. The first two indicate the total number of rows and columns. 
 * The last parameter indicates the index of the subplot.
 

```python
# Compute the x and y coordinates for points on sine and cosine curves

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1) # The three parameters: rows, columns, index of figure

# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot as active, and produce the second plot.
plt.subplot(2, 1, 2) # The three parameters: rows, columns, index of figure
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()

# Note: here the two subplots are arranged vertically
```

```python
# Two subplots arragned horizontally

x = np.linspace(0, 5, 10)
y = x ** 2

plt.subplot(1,2,1)
plt.plot(x, y, 'r--')

plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')

plt.show()
```

# Pie chart

```python
# Pie chart, where the slices are ordered and plotted counter-clockwise:

labels = 'Maths', 'Biology', 'Chemistry', 'Physics'

data = [15, 30, 45, 10] # percentile distribution
explode = (0, 0.1, 0, 0)  # only "explode" by 0.1 (degree of 'explosion') the 2nd slice called Biology

plt.pie(data, explode=explode, labels=labels,autopct='%1.1f%%',startangle=90)
plt.show()

# autopct='%1.1f%% allows the proportion of numbers appear in the respectives slices
```

# Bar plots

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5'] # We define five groups 
                                        
    
men_means = [20, 34, 30, 35, 27]    # each group corresponds to a value
women_means = [25, 37, 34, 27, 25]  # each group corresponds to a value

x = np.arange(len(labels)) # the location of the labels

width = 0.35 # the width of the bars
rects1 = plt.bar(x - width/2, men_means, width,label='Men')
rects2 = plt.bar(x + width/2, women_means, width,label='Women')

plt.legend()
plt.show()
```

# Figure size and resolution

You can specify the resolution, the size, and the aspect ratio of a plot

 * The example below creates an 800*600, 100dpi resolution image.
  

```python
x = np.linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(figsize=(8,6),dpi=100)

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')

plt.show()
```

# Using the subplot layout manager

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
for ax in axes.reshape(-1):  # axes returned as a 2x3 numpy array, reshaped into a vector
    ax.plot(x,y,'r')
    ax.set_ylabel('y axis')
    ax.set_xlabel('x axis')
    
fig.tight_layout()            # adjusts spacing to avoid plots overlapping labels
```

# 3D figures

 * 3D figures require another library to be imported 
 * mpl_toolkits.mplot3d library
 

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize=(8,4))

# prepare 3D data
axis=np.linspace(-5,5,100)
(X,Y) = np.meshgrid(axis,axis)    # outputs matrices X and Y storing grid of pixel coordinates
Z = np.cos(X) + np.sin(Y)        # perform some operation on the coordinate values

ax1 = fig.add_subplot(1,2,1, projection='3d') # generate an Axes3D
im = ax1.plot_surface(X,Y,Z, cmap=plt.cm.gray)


ax2 = fig.add_subplot(1,2,2, projection='3d') # generate an Axes3D
im = ax2.plot_wireframe(X,Y,Z, rstride=10, cstride=10)
```
