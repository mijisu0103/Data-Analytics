## 1.3 Conditioning and loops

* The if→Then→Else is a conditional statement that executes its sub-code only if the provided condition is found to be true. The else statement is optional.
* The for-loop specifies iteration, which allows code to be executed repeatedly.

* The while loop is used to execute a specific block of code for an unknown number of times, until a condition is met.

* Go through the example codes below and explore the output.

Note that Python, unlike most other languages, uses indentation (by default in sets of four spaces) to indicate blocks of code; e.g., the lines of code executed in a for-loop. It also does not use any character (e.g., ";" in many languages) to terminate a code statement.


```python
# if- elif- else
grade = 50
if grade >= 70:
    print("1st class")
elif grade >= 40:
    print("Passed")
else:
    print("Failed !")
```

```python
# for loop
data=[1,2,3,4,5]
sum=0
for item in data:
    sum += item
# loop ends here
print (sum)
```

```python
# while loop
data = [1,2,3,4,5]
n = 5
i =0
sum=0
while i < n:
    sum += data[i]
    i += 1
# loop ends here
print(sum)
```

You can also use list comprehension which represents a compact way for creating a list from a string or another list. Some examples are shown below.

```python
nums = [0, 1, 2, 3, 4]
squares = [ ]
for x in nums:
    squares.append(x ** 2)
print(squares)              # Prints [0, 1, 4, 9, 16]
```

```python
squares = [x ** 2 for x in nums]   # Using list comprehension (compact form)
print(squares)                     # Prints [0, 1, 4, 9, 16]
```

```python
even_squares = [x ** 2 for x in nums if x % 2 == 0] 
print(even_squares)                # Prints "[0, 4, 16]"
```

### 1.4.3 Tuples

* Tuples represent immutable lists. This means that, once created, they cannot be modified.
* They are useful when implementing larger projects, and eliminates the risk of incorrectly modifying a list that is not meant to be modified.
* Some common methods used with tuples are shown below.

```python
x = (1, 2)         # note ( ) is used for tuples and [ ] for lists
print(type(x))     # prints <class ‘tuple’>
```

```python
x = 1, 2           # the brackets are not strictly necessary
print(type(x))     # prints <class ‘tuple’>
```

```python
print (x[0])       # prints 1
```

```python
# x[0] = 5           # Returns an error since tuples are immutable!
```

```python
pos = (10, 20, 30)
(x, y, z) = pos       # 'unpack' tuple into separate variables
pos1 = (10, 20, 30) 
pos2 = (10, 25, 30)
pos1 == pos2          # Prints True if all elements equal
```

```python
x, y = 10, 15
x, y = y, x           # Can swap variables with a single line!
print (x, y)          # Prints 15 10
```

### 1.4.4. Dictionaries

* A dictionary maps a unique key to each value, and are mutable.
* Duplicate keys are not allowed. Duplicate values are allowed.
* Some common methods that are used with dictionaries are shown below.
* For more details, you can refer to the documentation available here:
https://docs.python.org/3.5/library/stdtypes.html#dict 

```python
office = {'Dan': 104, 'John':146,'Chris':245}
print(type(office))         # Prints <class ‘dict’>
```

```python
print(office['Dan'])       # look up a value for a given key; prints 104
```

```python
# print(office['Jose'])     # throws exception if key doesn't exist 
```

```python
office['Jose'] = 282     # add a new key-value pair 
print(office)            # prints all content with Jose added as last entry
```

```python
print(office.keys())     # return the list of keys 
```

```python
print(office.values())   # return the list of values 
```

```python
del office['John']      # remove an element from a dictionary 
print('Dan' in office)  # Prints true if the key exists 
```

```python
d = {'person': 2, 'cat': 4, 'spider': 8}
for animal in d:
    legs = d[animal]                          # Prints "A person has 2 legs", "A cat has 4 legs",
    print('A %s has %d legs' % (animal,legs)) # and "A spider has 8 legs".
```

Dictionary comprehension is similar to list comprehension, and enables us to easily construct dictionaries. For example:

```python
nums = [0, 1, 2, 3, 4]
even_num_to_square = {x: x ** 2 for x in nums if x % 2 == 0}
print(even_num_to_square) # Prints "{0: 0, 2: 4, 4: 16}
```

## 1.5 Functions

You can write our own functions using the keyword ‘def’. The general syntax for a function is:

* def function_name(parameters):
   * body of the function

Example 1: returning the sum of the elements in a list by calling a function.

```python
def sum_list(data): # Returns the sum of a list of numbers
    sum = 0
    for x in data:
        sum += x
    return sum

# Calling the above function:
result = sum_list([1,2,3,4,5,6,7,8])
print(result)
```

Example 2: returning multiple values.

```python
def max_min(data): # Returns min, max from the list 
    min, max = data[0], data[0]
    for item in data:
        if item < min:
            min = item
        elif item > max:
            max = item
    return max, min

# This function returns two values that will be returned as a tuple !
# Calling the function:
max, min = max_min([1,7,2,9,12,-7])
print("Max is " + str(max) + " and min is "+ str(min))
```

* When we call a function we need to specify its parameters (or arguments).
* A function involves parameters that have default values and parameters that do not have default values.
* Parameters that do not have a default value assingned must be specified when calling the function, while default parameters are optional; i.e., if not specified, their default value is assumed, as shown in Example 3 below.

```python
# a function with no default parameters
def compute_sum(a,b):
    return a+b

# Calling the function using positional parameters: 
ans = compute_sum(10,20)
print(ans)
```

```python
ans = compute_sum(b=20,a=10)
print(ans)
```

```python
# A function with some default and non-default parameters
def compute_sum(a, b, addC=False, c=99):
    if addC:
        return a+b+c 
    else:
        return a+b

# Calling the function:
ans = compute_sum(10,20,True)
print(ans)
```

```python
ans = compute_sum(b=10,a=20,addC=True,c=200)
print(ans)
```
